### Assignment 13: List Reversal

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.



```python
def reverse(lst):
    return lst[::-1]

originallist = list(map(int,input().split()))
reversed_list = reverse(originallist)
print(f"original list : {originallist}")
print(f"reversed list : {reversed_list}")
```

### Assignment 14: List Rotation

Write a function that rotates a list by n positions. Print the original and rotated lists.



```python
def rotate(lst,n):
    return lst[n:]+lst[:n]

original = list(map(int,input().split()))
n=int(input("Enter place where u wants to rotate"))
rotated = rotate(original,n)
print(f"original list: {original}")
print(f"rotated list : {rotated}")
```

### Assignment 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def uniquelst(lst1,lst2):
    lst1=set(lst1)
    lst2=set(lst2)
    bothlst = list(lst1.intersection(lst2))
    return bothlst

lst1=[1,2,3,4,5]
lst2=[3,4,5,6,7]
result = uniquelst(lst1,lst2)
print(f"first list : {lst1}")
print(f"second list : {lst2}")
print(f"new list with element present in both list : {result}")
```
