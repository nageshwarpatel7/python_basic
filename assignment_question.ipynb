**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Your code here
#by inbuild method
n=[6,8,2,4,1,9]
n.sort()
print(n)

#by sorting
n=[6,8,2,4,1,9]
for i in range(len(n)):
    for j in range(i+1,len(n)):
        if n[i]>n[j]:
            n[i],n[j]=n[j],n[i]
print(n)
```

```python
def generate_power_set(arr):
    n = len(arr)
    subsets = []

    # Generate all possible binary strings of length n
    for i in range(1 << n):
        subset = []
        for j in range(n):
            if i & (1 << j):
                subset.append(arr[j])
        subsets.append(subset)

    # Sort the subsets lexicographically
    subsets.sort()

    return subsets

# Example usage
input_array = [1, 2, 3]
result = generate_power_set(input_array)
for subset in result:
    print(subset)
```
