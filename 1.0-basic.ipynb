Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured in programs in a language. In simpler terms, syntax is about the correct arrangement of words and symbols in code.

Semantics refers to the meaning or the interpretation of the symbols, characters and commands in a language. It is about the what the code is supposed to do when it runs.

```python
## Basic syntax rules in python 
## Case sensitivity- Python is case sensitive 

name = "Krish"
Name = "Naik"

print(name)
print(Name)
```

```python
## Indentation 
## Python uses indentation to define a blocks of code. Consistent use of spaces (commonly 4) or a tab is required.

age = 32
if age>30:
    print(age)
    
print(age)
```

```python
## Line Continuation 
## Use a backslash(\) to continue a statement to the next line

total = 1+2+3+4+5+6+7+\
    4+5+6
    
print(total)
```

```python
## Multiple Statements on a single line
x=5;y=10;z=x+y
print(z)
```
